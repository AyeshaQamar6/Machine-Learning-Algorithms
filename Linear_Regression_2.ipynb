## Predicting cancer and its terms
### Using Linear Regression Algorithm

```python
# Importing Libraries
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Loading Dataset by sklearn dataset
cancer = datasets.load_breast_cancer()

# Viewing cancer data
print(cancer.data)
```

```python
# Viewing keys in cancer dataframe
print(cancer.keys())
```

```python
# Viewing cancer dataset information
print(cancer.DESCR)
```

```python
# Convert elements into list of list using numpy array
cancer = datasets.load_breast_cancer()
cancer_x = cancer.data[:, np.newaxis, 6]
print(cancer_x)
```

```python
# Slicing for training and testing

# Features for x-axis
cancer_x_train = cancer_x[:-10]   #(Last 10 values)
cancer_x_test = cancer_x[-15:]   #(Starting 15 values)
```

```python
# Label for y-axis
cancer_y_train = cancer.target[:-10]   #(Last 10 values)
cancer_y_test = cancer.target[-15:]   #(Starting 15 values) #(Actual Value)
```

```python
'''cancer_x = np.array(((1), (2), (3)))

cancer_x_test = cancer_x'''
```

```python
'''cancer_y_test = np.array((3, 2, 4))'''
```

```python
# For making Linear Model
model = linear_model.LinearRegression()
```

```python
# Fitting the data

# Making the model (Making a linear line)
model.fit(cancer_x_train, cancer_y_train)
```

```python
# Predicting the model

# Testing the model
cancer_y_predict = model.predict(cancer_x_test)  # (Predicted Value)
```

```python
# Calculating mean squared error
# Sum of(predicted-actual value)^2 / (Average of error or Average of sum of error)

print("Mean Squared Error: ", mean_squared_error(cancer_y_test , cancer_y_predict))
```

```python
# Viewing weights of dataset

print("Weights: ", model.coef_)
```

```python
# Viewing intercept of dataset

print("Intercept: ", model.intercept_)
```

#### Plotting

```python
# Scatter Plot

plt.scatter(cancer_x_test, cancer_y_test)
plt.show()
```

```python
# Prediction plot

plt.scatter(cancer_x_test, cancer_y_test)
plt.plot(cancer_x_test, cancer_y_predict)
plt.show()
```
